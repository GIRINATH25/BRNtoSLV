```
def build_document(row):
    parts = [
        f"Table: {row['source_tablename']}",
        f"Column: {row['source_columnname']}",
        f"Type: {row['datatype']}"
    ]
    if 'description' in row and pd.notna(row['description']):
        parts.append(f"Description: {row['description']}")
    return " | ".join(parts)

df["document"] = df.apply(build_document, axis=1)
```

```
import chromadb
from chromadb.utils import embedding_functions
```

```
embedding_fn = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)
chroma_client = chromadb.Client()
```

```
collection = chroma_client.create_collection(
    name="metadata_lookup",
    embedding_function=embedding_fn
)
```

```
collection.add(
    documents=df["document"].tolist(),
    metadatas=df[["source_tablename", "source_columnname", "datatype"]].to_dict(orient="records"),
    ids=[f"row_{i}" for i in range(len(df))]
)
```

```
def search_metadata(user_query, top_k=None):
    n_results = collection.count() if top_k is None else top_k

    results = collection.query(
        query_texts=[user_query],
        n_results=n_results
    )

    distances = results["distances"][0]
    documents = results["documents"][0]
    metadatas = results["metadatas"][0]

    # Sort by distance
    sorted_data = sorted(
        zip(distances, documents, metadatas),
        key=lambda x: x[0]
    )

    sorted_distances = [x[0] for x in sorted_data]

    # Detect largest gap
    gaps = [
        sorted_distances[i+1] - sorted_distances[i]
        for i in range(len(sorted_distances)-1)
    ]

    if not gaps:
        return results

    split_index = gaps.index(max(gaps)) + 1

    # Keep only strong cluster
    strong_matches = sorted_data[:split_index]

    filtered = {
        "documents": [[x[1] for x in strong_matches]],
        "metadatas": [[x[2] for x in strong_matches]],
        "distances": [[x[0] for x in strong_matches]],
    }

    return filtered
```

```
from collections import defaultdict

def rank_tables(results):
    table_scores = defaultdict(list)

    for doc, meta, dist in zip(
        results["documents"][0],
        results["metadatas"][0],
        results["distances"][0]
    ):
        table_scores[meta["source_tablename"]].append({
            "column": meta["source_columnname"],
            "score": dist,
            "doc": doc
        })

    # Sort tables by best (lowest distance)
    ranked_tables = sorted(
        table_scores.items(),
        key=lambda x: min(col["score"] for col in x[1])
    )

    return ranked_tables
```

```
query = "finance"

results = search_metadata(query)
ranked_tables = rank_tables(results)

print(results,'res ******')
print(ranked_tables,'rank *****')
```

```
for table, columns in ranked_tables:
    print(f"\nTable: {table}")
    for col in columns:
        print(f"  - Column: {col['column']} | Score: {col['score']:.4f}")
```
